### Data Preparation


1) Rename column <b>Num cases </b> to <b>Num Cases</b> in raw_data1 and raw_data2 to avoid confusion. <br><br>
2) Drop Column(Estimated Onset Date) and (Backup Notes) from raw_data1 and raw_data2 : No use and not present in further data sets.


```python
df1.shape #21 Column
df2.shape #21 Column
df3.shape #20 Column
df4.shape #20 Column
df5.shape #20 Column
df6.shape #20 Column
df7.shape #20 Column
df8.shape #20 Column
df9.shape #20 Column
df10.shape #20 Column
df11.shape #20 Column
df12.shape #20 Column
df13.shape #20 Column
df14.shape #20 Column
```

```python
#rename column and drop column
df1.rename(columns={"Num cases": "Num Cases"}, inplace=True)
df2.rename(columns={"Num cases": "Num Cases"}, inplace=True )

df1.drop(['Estimated Onset Date', 'Backup Notes'], axis=1,inplace=True)
df2.drop(['Estimated Onset Date', 'Backup Notes'], axis=1,inplace=True)
```

### Statewise Total Confirmed || Recovered || Deceased Cases

```python
#Hospitalized
df_confirmed=df_new[df_new['Current Status']=='Hospitalized'].copy()
df_confirmed.head(3)
df_confirmed.rename(columns={"Num Cases": "Confirmed_Cases"}, inplace=True)

#Recovered
df_recovered=df_new[df_new['Current Status']=='Recovered'].copy()
df_recovered.head(3)
df_recovered.rename(columns={"Num Cases": "Recovered_Cases"}, inplace=True)

#Deceased
df_deceased=df_new[df_new['Current Status']=='Deceased'].copy()
df_deceased.head(3)
df_deceased.rename(columns={"Num Cases": "Deceased_Cases"}, inplace=True)

```

```python
df_confirmed.head()
```

```python
confirmed_statewise_count = df_confirmed.groupby("Detected State")[["Confirmed_Cases"]].sum().sort_values(by='Detected State', ascending=True)
recovered_statewise_count = df_recovered.groupby("Detected State")[["Recovered_Cases"]].sum().sort_values(by='Detected State', ascending=True)
deceased_statewise_count = df_deceased.groupby("Detected State")[["Deceased_Cases"]].sum().sort_values(by='Detected State', ascending=True)
```

```python
merged_df = pd.concat([confirmed_statewise_count, recovered_statewise_count,deceased_statewise_count],axis=1)
merged_df
merged_df.reset_index(inplace=True)
```

```python
merged_df.rename(columns={"index": "State"}, inplace=True)
```

```python

merged_df.dropna(subset=['Recovered_Cases'],inplace = True)
merged_df["Deceased_Cases"].fillna(0, inplace = True)
merged_df["Recovered_Cases"] = merged_df["Recovered_Cases"].astype("int")
merged_df["Deceased_Cases"] = merged_df["Deceased_Cases"].astype("int")
merged_df['Active_Cases'] = merged_df['Confirmed_Cases'] - (merged_df['Recovered_Cases'] + merged_df['Deceased_Cases'])
```

```python
merged_df.style.background_gradient(cmap='Reds')
```

```python
plt.rcParams['figure.figsize'] = 15,7
merged_df.plot(kind="bar",x="State",y="Confirmed_Cases")
```

```python
plt.rcParams['figure.figsize'] = 15,7
merged_df.plot(kind="bar",x="State",y="Active_Cases")
```

```python
fig = plt.figure(figsize=(20,10))
axes = fig.add_axes([0,0,1,1])
axes.bar(merged_df["State"],merged_df["Active_Cases"])
axes.set_title("Covid 19 Cases in India")
axes.set_xlabel("Name of State/UT")
axes.set_ylabel("Total Cases")
```

```python

dataset=pd.read_csv('df_confirmed_date_count.csv')
```

```python
X=dataset['Days']
y=dataset['Num Cases']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.90, random_state=42)
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
prediction=lm.predict(X_test)
prediction
```

```python
plt.scatter(prediction,y_test)
```

```python

sns.distplot((y_test-prediction),bins=50);
```

```python
from sklearn import metrics
import numpy as np
print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```
